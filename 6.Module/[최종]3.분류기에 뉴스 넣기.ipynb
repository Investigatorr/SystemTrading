# 날짜 함수(2일전 날짜 파싱용)

```python
from datetime import datetime, timedelta, date


def getday(n):
    n_day = timedelta(days=n)
    res =  date.today() - n_day
    end = res.strftime("%Y.%m.%d")
    return end


def gettoday():
    today = datetime.today().strftime("%Y.%m.%d")
    return today
```
