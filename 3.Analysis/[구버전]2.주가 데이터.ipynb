## normalize data

```python
#normalize by row
df = df.apply(lambda x: (x- x.mean()) / x.std(), axis = 1)
df.head()
```

```python
#normalize by entire data
df = df.apply(lambda x: (x- df.mean().mean())/df.std().std(), axis=0)
df.head()
```

```python
#내일 주가 수익률 
df['diff'] = (df['Close'].shift(-1) - df['Close']) / df['Close']
df.dropna(inplace=True)
#df['diff'].head()
```

```python
#주가 상승률 구간 나눔
idx = [meanC - stdC * i for i in range(10, 0, -1) if meanC - stdC * i >= df['diff'].min()] + [meanC + stdC * i for i in range(10) if meanC + stdC * i <= df['diff'].max()]
np.array(idx)
```

```python
df['cls'] = df['diff'].apply(lambda x:np.abs(np.array(idx) - x).argmin())
```

```python
#260 -> 229
df.dropna(inplace=True)
```

```python
'''
    data: 종가 기준으로 나눈 구간(cls), 수익률(diff) 컬럼을 제외한 data
    target: cls
    
    t_data: 데이터 중 100개만 샘플로 뽑아서.. (X)
    t_target: ''                              (y)
'''
data = df.drop(['diff','cls'],axis=1)
target = df['cls']

#training data: 첫날부터 100개
train_df = df.iloc[:100] #df.sample(100)
train_data = train_df.drop(['diff','cls'],axis=1)
train_target = train_df['cls']

#test data: training 100개를 제외한 데이터 
# test_df = df.drop(train_df.index)
test_df = df.iloc[len(train_df):len(train_df)+5]#.to_frame().T
test_data = test_df.drop(['diff','cls'],axis=1)
test_target = test_df['cls']
print(train_df.iloc[-1], test_df)
```

# SVM

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score


clf = SVC(class_weight='balanced', gamma=0.11125, C=0.2275, kernel='rbf')

clf.fit(train_data, train_target) 

y_pred = clf.predict(test_data)
y_true = test_target

print(y_pred)
print(accuracy_score(y_true, y_pred))

confusion_matrix(y_true, y_pred)
```
