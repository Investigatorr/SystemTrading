# 데이터 임포팅

```python
# 뉴스데이터 임포팅
news = pd.read_csv('samsung_pre.csv', encoding = 'utf-8 sig')
news.head()
```

```python
del news['Unnamed: 0']
news.head()
```

```python
# 원본데이터는 유지하면서 news2라는 더미 생성
news2 = news.copy()
news2.head()
```

```python
# 두 df를 합치기 전 key값이 되는 칼럼 이름 맞추기
news2.rename(columns={'TIME': 'Date'}, inplace=True)
```

```python
# 칼럼 삭제
del news2['TITLE']
del news2['BODY']
```

```python
# 칼럼 순서 바꾸기
news2 = news2[['title+body', 'NOUNS','Date']]
```

# pandas에서 날짜를 기준(group by)으로 list형태의 명사 합치기

```python
# 백업을 위해 새로운 news2 df 만들기
news3 = news2.copy()
```

# !!!!주가데이터 임포팅

```python
# 1,3,5,10 일간의 diff와 이동평균이 필요
stock = pd.read_csv('stockprice.csv', encoding = 'utf-8 sig')
stock.head()
```

```python
# 칼럼 합치기 (full outer join) / 한쪽이 빈 값은 Null값으로 채워짐
# merge에 대한 자세한 설명   https://www.shanelynn.ie/merge-join-dataframes-python-pandas-index-1/

df = pd.merge(news3, stock, on=['Date'], how='outer')

print('뉴스 데이터 : {}'.format(news3.shape))
print('주가 데이터 : {}'.format(stock.shape))
print('합친 데이터 : {}'.format(df.shape))
```

```python
df.count()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.tail()
```

```python
df['Diff1'] = df['Close'] - df['day1C']
df['Diff3'] = df['Close'] - df['day3C']
df['Diff5'] = df['Close'] - df['day5C']
df['Diff10'] = df['Close'] - df['day10C']
```

### diff, 즉 y값들을 상승/하락률에 따라 범주화 시키기
- https://datascienceschool.net/view-notebook/aa62265f02fc429aa636ef343c3b1fda/
- 0~1: + 
- 1~3: ++
- 3~5: +++
- 5~10: ++++
- 10~20: +++++
- 20~? : '가즈아ㅏㅏ' / 'Do It JohnBer'

```python
# 상승률 / 하락률 구해 퍼센트로 치환 (단순 나눈기 + round함수로 소수2자리까지만)
df['Diff1_per'] = round(df['Diff1'] / df['Close'] * 100, 2)
df['Diff3_per'] = round(df['Diff3'] / df['Close'] * 100, 2)
df['Diff5_per'] = round(df['Diff5'] / df['Close'] * 100, 2)
df['Diff10_per'] = round(df['Diff10'] / df['Close'] * 100, 2)
```

```python
# 카테고리화 시키기
bins = [-100, -20, -10, -5, -3, -1, 0, 1, 3, 5, 10, 20, 100]
labels = ['Do it Johnber', '-5등급', '-4등급', '-3등급', '-2등급', '-1등급', '+1등급', '+2등급', '+3등급', '+4등급', '5등급', '가즈아ㅏㅏ']
df['Diff1_clf'] = pd.cut(df["Diff1_per"], bins, labels = labels)
df['Diff3_clf'] = pd.cut(df["Diff3_per"], bins, labels = labels)
df['Diff5_clf'] = pd.cut(df["Diff5_per"], bins, labels = labels)
df['Diff10_clf'] = pd.cut(df["Diff10_per"], bins, labels = labels)
```
